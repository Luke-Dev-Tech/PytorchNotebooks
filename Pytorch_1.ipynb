torch. rand() generates numbers evenly spread between 0 and 1 . torch. randn() generates numbers based on the bell curve, centered at 0 with most values falling between -3 and 3

```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x_randn = torch.randn(size=(224,224,3))
print("x with randn ==> ", x_randn)
print("===================================")

x = torch.rand(2,10,2)
print("===================================")
print('x with rand ==> ',x)

x_original = torch.rand(size=(224,224,3)) # [height, width, color_channels]
x_original.shape
print("===================================")
print("x with rand and size parameter ==> ",x_original)
print(type(x_original))
```

# Pytorch tensors and NumPy


Numpy is a popular scientific Python numerical computing library and because of this, Pythorch has functionality to interact with it.

* Data in NumPy, want in PyTorch tensor -> 'torch.from_numpy(ndarray)'


Pytorch tensor -> Numpy -> torch.Tensor.numpy()

```python
# Numpy array to tensor

import torch
import numpy as np


array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) # warning: when numpy --> pytorch (it reflects numpy's float64)
array, tensor
```

```python
array.dtype
```

```python
tensor.dtype
```

```python
torch.arange(1.0, 8.0).dtype # pytorch's default is float32
```

```python
# let's try changing the array
array = array + 1
array, tensor
# it doesn't change the tensor value
```

```python
# torch to numpy

tensor_ones = torch.ones(7)

numpy_tensor = tensor_ones.numpy()
tensor_ones, numpy_tensor

# they reflects each other.
```

```python
tensor_ones = tensor_ones + 1
tensor_ones, numpy_tensor
```

Pytorch Reproducbuility
(Random out of Random)

How neural network learns is
'start with random numbers --> tensor operations --> update random numbers to try and make them better representation of the data -> again --> again ...


To reduce the randomness in the neural networks and Pytorch comes the concpet of **random seed**

```python
torch.rand(3,3,3)
```

```python
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
Random_seed = 42
torch.manual_seed(Random_seed) # .manual seed, works only for one block if you are using a notebook
random_tensor_A = torch.rand(3,4)


torch.manual_seed(Random_seed)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Check for GPU access with Pytorch
# CUDA in PyTorch allows your code to run on the GPU instead of the CPU.
import torch
torch.cuda.is_available()
```

```python
import torch

# Create a tensor on the CPU
x = torch.randn(3, 3)

# Move it to the GPU (if available)
if torch.cuda.is_available():
    tensor_on_gpu = x.to('cuda' )
    print("Tensor is now on GPU:", tensor_on_gpu.device)
```

```python
x = torch.randn(3, 3, device='cuda')
x
```

```python
# Note Tensor only works with CPU
# so moving back to cpu
# but if tensor is on GPU , Can't transform it to NUMPY.

tensor_on_gpu.numpy() # device issue
```
