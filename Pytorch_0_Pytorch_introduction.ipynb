# Tensor

a tensor can be thought of as a generalization of a matrix.

A scalar is a 0-dimensional tensor (a single number).

A vector is a 1-dimensional tensor (a list of numbers).

A matrix is a 2-dimensional tensor (a grid of numbers).

A tensor can have any number of dimensions (a collection of grids).

## Dimension

--> How many Axis does the tensor has.

## Shape

--> How many element are there in the tensor

Example,

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x.ndim)   # Number of dimensions → 2

print(x.shape)  # Shape of tensor → torch.Size([2, 3])

### Reshaping, stacking, squezzing and unsquezzing tensors


* Reshaping - reshapes an input tensor to a defined shape

* View - return a view of an input tensor of certain shape but keep the same memory as the original tensor

* Stacking - combine multiple tensors on the top of each other (vstack) or side by side (hstack).

* squeeze - removes all '1' dimensions from a tensor.

* Unsqueeze - add a '1' dimension to a target tensor.


* Permute - Return a view of the input with dimension permuted (swapped) in a certain way




````python
x = torch.arange(1.,10.,1)
x, x.shape
````

````python
# When reshaping, you need to make sure it is
# compatible with what you have.

x_reshaped = x.reshape(1,10) # you can't just add extra column
x_reshaped, x_reshaped.shape
````

````python
x_reshaped = x.reshape(2,9) # you can't just add extra row
x_reshaped, x_reshaped.shape
````

### torch.permute
rearrange the dimentions of a target tensor in a specified order

````python
# return the view of the original tensor input with its dimensions permuted (rearanged)
````

````python
x = torch.randn(2,3,5)
x.size(), x.ndim
````

````python
torch.permute(x, (2,0,1)).size() # (second D, 0 D, first D) <-- arragned accordingly
````

````python
x_original = torch.rand(size=(224,224,3)) # [height, width, color_channels]
x_original.shape
print(x_original)
print(type(x_original))
````

````python
# Permute the original tensor to rearrange the axis

print('Permute to (0,2,1) --> (224,3,224)')
x_permuted_0_2_1 = torch.permute(x_original, (0,2,1))
print(x_permuted_0_2_1)
print(x_permuted_0_2_1.shape)
print(x_permuted_0_2_1.ndim)
print('__________________________________')
print(torch.permute(x_original, (0,2,1)))
print('__________________________________')
print('Permute to (2,0,1) --> (3,224,224)')
x_permuted = torch.permute(x_original, (2,0,1))
print(x_permuted)
````
